```python
from random import randrange
class Player(object):
    def __init__(self, symbol, isAI):
        self.symbol = symbol
        self.isAI = isAI
        
    def play(self, tictactoe_Board):
        if self.isAI:
            print('Computer Turn...')
            global count
            depth, alpha, beta, isMax = 9-count, -10, 10, True
            value, place = self.alphaBetaPrunning(tictactoe_Board,depth, alpha, beta, isMax)
            print('Computer plays at:', place+1)
        else:
            print('Your Turn...')
            place = int(input('Enter a number from 1 to 9: ')) - 1
            while self.isInvalid(tictactoe_Board, place):
                place = int(input('Enter again: ')) - 1
        tictactoe_Board[place] = self.symbol
            
    def alphaBetaPrunning(self, tictactoe_Board, depth, alpha, beta,isMaxPlayer):
        value = checkGameOver(tictactoe_Board, player)
        if value != 0:
            return value, -1
        if depth == 0:
            return value, -1
        if isMaxPlayer:
            value = -10
            listOfNumbers = list(range(9))
            while listOfNumbers:
                random = listOfNumbers.pop(randrange(len(listOfNumbers)))
                if tictactoe_Board[random] == 'b':
                    tictactoe_Board[random] = player[1].symbol
                    temp_value = self.alphaBetaPrunning(tictactoe_Board, depth-1, alpha, beta, False)[0]
                    if temp_value > value:
                        value = temp_value
                        place = random
                    alpha = max(alpha, value)
                    tictactoe_Board[random] = 'b'
                    if alpha >= beta:
                        break
        else:
            value = 10
            listOfNumbers = list(range(9))
            while listOfNumbers:
                random = listOfNumbers.pop(randrange(len(listOfNumbers)))
                if tictactoe_Board[random] == 'b':
                    tictactoe_Board[random] = player[0].symbol
                    temp_value = self.alphaBetaPrunning(tictactoe_Board, depth-1, alpha, beta, True)[0]
                    if temp_value < value:
                        value = temp_value
                        place = random
                    beta = min(beta, value)
                    tictactoe_Board[random] = 'b'
                    if beta <= alpha:
                        break
        return value, place
    
    def isInvalid(self, tictactoe_Board, place):
        if place not in list(range(9)):
            print('ERROR: Invalid input. Number not from 1 to 9.')
            return True
        if tictactoe_Board[place] != 'b':
            print('ERROR: Invalid input. Place already marked.')
            return True
        return False
    
    def isSymbolX(self):
        return self.symbol == 'X'

def myMsg():
    msg = 'Welcome to the Tic Tac Toe game. You will play the classic '\
    + 'Tic Tac Toe game against the hardest AI. Rules of the game are␣simple: '\
    + 'Try to mark your X\'s (or O\'s) three in row or column or diagonally'\
    + 'to win the game. To mark on a particular place **ENTER A NUMBER '\
    + 'FROM 1 TO 9 WITH AN EMPTY PLACE (MARKED AS b).**\nAll the Best! :)'
    print(msg)
    
def initialize():
    global tictactoe_Board, player, turn, count
    tictactoe_Board, count = list('b'*9), 0
    if input('Do you want to play first? (yes/no)\n')[0].lower() == 'y':
        player, turn = [Player('X', False), Player('O', True)], 0
    else:
        player, turn = [Player('O', False), Player('X', True)], 1
    
def drawBoard(tictactoe_Board):
    for i in range(9):
        print(tictactoe_Board[i], '\n'*((i%3)-1), end='')
    
def displayWinMsg(whoWon):
    if whoWon == 1:
        print('Sorry. You lose. Better luck next time.')
    elif whoWon == -1:
        print('Congratulations! You Win!')
    else:
        print('It\'s a Draw.')
    
def checkGameOver(tictactoe_Board, player):
    x = 3
    symbolToNumber, myGame = {}, [[0,0,0],[0,0,0],[0,0,0]]
    symbolToNumber['X'] = 2*player[1].isSymbolX() - 1
    symbolToNumber['O'] = 2*player[0].isSymbolX() - 1
    symbolToNumber['b'] = 0
    for i in range(x):
        for j in range(3):
            myGame[i][j] = symbolToNumber[tictactoe_Board[i*3+j]]
    for i in range(x):
        rowcount, colcount = 0, 0
        for j in range(3):
            rowcount += myGame[i][j]
            colcount += myGame[j][i]
        if (rowcount == -3 or colcount == -3):
            return -1
        if (rowcount == 3 or colcount == 3):
            return 1
    if myGame[1][1] == -1:
        if (myGame[0][0] == -1 and myGame[2][2] == -1) or (myGame[0][2] == -1 and myGame[2][0] == -1):
            return -1
    if myGame[1][1] == 1:
        if (myGame[0][0] == 1 and myGame[2][2] == 1) or (myGame[0][2] == 1 and myGame[2][0] == 1):
            return 1
    return 0

if __name__ == '__main__':
    myMsg()
    optForChoice = 'y'
    while optForChoice == 'y':
        initialize()
        whoWon = False
        while not whoWon:
            drawBoard(tictactoe_Board)
            player[turn].play(tictactoe_Board)
            whoWon = checkGameOver(tictactoe_Board, player)
            turn = 1 - turn
            count += 1
            if count == 9:
                break
        drawBoard(tictactoe_Board)
        displayWinMsg(whoWon)
        optForChoice = input('Play Again? (yes/no)\n')[0].lower()
```
